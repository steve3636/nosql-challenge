### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Define the center of the search area (latitude and longitude)
latitude = 51.5074  # Example latitude for London
longitude = -0.1278  # Example longitude for London

# Define the degree search range (0.01 degrees on either side of the center)
degree_search = 0.01

# Construct the query to match establishments
query = {
    'geocode.latitude': {
        '$gte': latitude - degree_search,
        '$lte': latitude + degree_search
    },
    'geocode.longitude': {
        '$gte': longitude - degree_search,
        '$lte': longitude + degree_search
    },
    'RatingValue': 5
}

# Define the sort stage to sort by hygiene score
sort = {
    'scores.Hygiene': 1  # Sort in ascending order (lowest hygiene score first)
}

# Create the aggregation pipeline with match, sort, and limit stages
pipeline = [
    {
        '$match': query
    },
    {
        '$sort': sort
    },
    {
        '$limit': 5
    }
]

# Execute the aggregation pipeline
results = list(establishments.aggregate(pipeline))

# Print the results
for result in results:
    print(f"BusinessName: {result['BusinessName']}")
    print(f"Hygiene Score: {result['scores']['Hygiene']}")
```

```python
# Execute the aggregation pipeline and store the results
results = list(establishments.aggregate(pipeline))

# Convert the results to a Pandas DataFrame
df = pd.DataFrame(results)

# Display the DataFrame
df
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

# Print the number of documents in the result

# Print the first 10 results
```

```python
# Define the aggregation pipeline
pipeline = [
    {
        '$match': {
            'Hygiene': 0
        }
    },
    {
        '$group': {
            '_id': '$LocalAuthorityName',
            'count': {'$sum': 1}
        }
    },
    {
        '$sort': {
            'count': -1  # Sort by count in descending order (highest to lowest)
        }
    }
]

# Execute the aggregation pipeline
results = establishments.aggregate(pipeline)

# Print the number of documents in the result
count = establishments.count_documents({'Hygiene': 0})
print(f"Number of documents with Hygiene score of 0: {count}")

# Print the first 10 results
for i, result in enumerate(results):
    if i < 10:
        local_authority_name = result['_id']
        establishment_count = result['count']
        print(f"Local Authority Name: {local_authority_name}")
        print(f"Number of Establishments with Hygiene Score of 0: {establishment_count}")
        print("\n" + "=" * 50 + "\n")
```

```python
# Convert the result to a Pandas DataFrame

# Display the number of rows in the DataFrame

# Display the first 10 rows of the DataFrame
```

```python
import pandas as pd

# Execute the aggregation pipeline and store the results in a list
pipeline_results = list(establishments.aggregate(pipeline))

# Convert the results to a Pandas DataFrame
df = pd.DataFrame(pipeline_results)

# Display the number of rows in the DataFrame
num_rows = len(df)
print(f"Number of rows in the DataFrame: {num_rows}")

# Display the first 10 rows of the DataFrame
print("First 10 rows of the DataFrame:")
print(df.head(10))
```
